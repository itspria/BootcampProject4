```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
csv_path = "../raw data/IMDB_movies.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
df.count()
```

```python
df_alt = df.drop(["budget","usa_gross_income", "worlwide_gross_income", "metascore"], axis=1)
df_alt.head()
```

```python
df_dropped = df_alt.dropna(how='any')
```

```python
df_dropped.count()
```

```python
df_dropped['date_published'] = pd.to_datetime(df_dropped['date_published'], format="%Y/%m/%d", errors='coerce')
```

```python
df_dropped['month_published'] = pd.DatetimeIndex(df_dropped['date_published']).month
```

```python
df_dropped.dtypes
```

```python
df_new = df_dropped.dropna(how='any')
```

```python
df_new['month_published'] = df_new['month_published'].astype(np.int64)
```

```python
df_new.head()
```

```python
df_new[['director', 'director2']] = df_new['director'].str.split(',', expand=True)
```

```python
df_2 = df_new.drop('director2', 1)
df_2.head()
```

```python
df_2[['writer', 'writer2']] = df_2['writer'].str.split(',', expand=True)
```

```python
df_2.head()
```

```python
df_3 = df_2.drop('writer2', 1)
df_3.head()
```

```python
df_3[['genre', 'genre_2', 'genre_3']] = df_3['genre'].str.split(',', expand=True)
```

```python
df_3[['language', 'language_2', 'language_3', 'langauge_4', 'lanaguage_5']] = df_3['language'].str.split(',', expand=True)
```

```python
df_3.head()
```

```python
df_4 = df_3.loc[df_3['country'] == 'USA']
```

```python
df_5 = df_4.drop('date_published', 1)
df_5.head()
```

```python
df_6 = df_5.drop('original_title', 1)
df_6.head(70)
```

```python
df_6.to_csv('movies.csv', index=False)
```
