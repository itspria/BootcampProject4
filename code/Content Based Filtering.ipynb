### Lets Recommend movies based on the genres, movies which are similar in type, Based on users last watched movie we recommend few similar movies having similar kind of genre.

```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
# File to Load
file_to_load = "../raw data/movies.csv"

# Read HDI File and store into Pandas data frame
data = pd.read_csv(file_to_load)

#display dataframe
data.head()
```

```python
data['genres'] = data['genres'].str.replace(r'|', ' ')
data
```

```python
genre_data=data[['title','genres']]
genre_data=genre_data.set_index('title')
genre_data
```

```python
#convert genre column to array
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(genre_data["genres"]).toarray()
X
```

```python
#Each row corresponds is a different movie
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(X) 
#Each row of matrix coressponds to similarity of a movie with all other movies
similarities
```

```python
# File to Load
file_to_load = "../raw data/ratings.csv"

# Read HDI File and store into Pandas data frame
ratings = pd.read_csv(file_to_load)

#display dataframe
ratings.head()
```

```python
output = data.loc[:,['movieId','title']]
output = output.join(pd.DataFrame(X))
output
```

```python
uid = 18 #For user 18 lets recommend movies based on his recent watched movie
time = ratings.loc[ratings["userId"]==uid,["movieId","timestamp"]]
latest_movieId_watched_by_user = time.sort_values(by="timestamp",ascending=False)["movieId"].values[0]
latest_movieId_watched_by_user
```

```python
movie_index = data.loc[data['movieId']==latest_movieId_watched_by_user,["title"]].index[0]
output.loc[output['movieId']==8798,:]
```

```python
#we need index but we are using id to find which row is crct in similarities matrix
movie_index = data.loc[data['movieId']==latest_movieId_watched_by_user,["title"]].index[0]
similarity_values = pd.Series(similarities[movie_index])
```

```python

#We converted list into series in order to preserve the actual indexes of dataset even after sorting
similarity_values.sort_values(ascending=False)
```

```python

similar_movie_indexes = list(similarity_values.sort_values(ascending=False).index)
```

```python
#Remove the already watched movie from index list
similar_movie_indexes.remove(movie_index)
```

```python
def get_movie_by_index(idx):
    return genre_data.index[idx]
def get_movie_by_id(mv_id):
    return data.loc[data['movieId']==mv_id,['title']].values[0][0]
get_movie_by_index(8899)
```

```python
print("Since u watched --->",get_movie_by_id(latest_movieId_watched_by_user),"<--- We recommend you")
for i in range(15):
    print(get_movie_by_index(similar_movie_indexes[i]))
```
