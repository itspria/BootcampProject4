```python
#import dependencies
import pandas as pd
import numpy as np
```

# Cleaning names.csv

```python
# File to Load
file_to_load = "../raw data/names.csv"

# Read HDI File and store into Pandas data frame
names_data = pd.read_csv(file_to_load)

#display dataframe
names_data.head()
```

```python
# dropping birth name column
names_data= names_data.drop(['birth_name'], axis=1)
names_data.head()
```

```python
#number of Null values in each column
names_data.isnull().sum(axis = 0)
```

```python
len(names_data.index)
```

```python
#get % of NaN values relative to total data set
null_values= names_data.isnull().sum(axis = 0)
perc_null = null_values/len(names_data.index)*100
perc_null
```

```python
# dropping columns with more than 50% Nan values
names_data= names_data.drop(columns=['height','birth_details','date_of_birth','place_of_birth','death_details','date_of_death','place_of_death','reason_of_death','spouses_string'])
names_data
```

```python
#checking data types
names_data.info()
```

# Cleaning title_principals.csv

```python
# File to Load
file_to_load = "../raw data/title_principals.csv"

# Read HDI File and store into Pandas data frame
tp_data = pd.read_csv(file_to_load)

#display dataframe
tp_data.head()
```

```python
#get % of NaN values relative to total data set
null_values= tp_data.isnull().sum(axis = 0)
perc_null = null_values/len(tp_data.index)*100
perc_null
```

```python
# dropping columns with more than 50% Nan values
tp_data= tp_data.drop(columns=['job','characters'])
tp_data
```

```python
#checking data types
tp_data.info()
```

# Join names and title principals tables

```python
tp_data.head()
```

```python
names_data.head()
```

```python
joined_names = pd.merge(names_data, tp_data)
joined_names.head()
```

```python
#Export joined file 
output_data_file = "../cleaned data/joined_names.csv"
joined_names.to_csv(output_data_file, index = False)
```

```python
#Export individual file 
output_data_file = "../cleaned data/names_data.csv"
names_data.to_csv(output_data_file, index = False)
```

```python
#Export individual file 
output_data_file = "../cleaned data/tp_data.csv"
tp_data.to_csv(output_data_file, index = False)
```
